```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x=np.linspace(-10,10,100)
y_sigmoid=sigmoid(x)

plt.figure(figsize=(10,6))
plt.plot(x,y_sigmoid,label="Sigmoid",color="blue")
plt.grid()
plt.title("Sigmoid activation function")
```

```python
def tanh(x):
    return np.tanh(x)

x=np.linspace(-10,10,100)
y_tanh=tanh(x)

plt.figure(figsize=(10,6))
plt.plot(x,y_tanh,label="tanh",color="red")
plt.grid()
plt.title("Tanh activation function")
```

```python
def relu(x):
    return np.maximum(0,x)

x=np.linspace(-10,10,100)
y_relu=relu(x)

plt.figure(figsize=(10,6))
plt.plot(x,y_relu,label="relu",color="pink")
plt.grid()
plt.title("Relu activation function")
```

```python
def leaky_relu(x,alpha=0.01):
    return np.where(x>0,x,alpha*x)#np.where(condition, value_if_true, value_if_false)

alpha=0.01
x=np.linspace(-10,10,100)
y_leaky_relu=leaky_relu(x,alpha)
plt.figure(figsize=(10,6))
plt.plot(x,y_leaky_relu,label="leaky relu",color="orange")
plt.grid()
plt.title("Leaky relu activation function")

```

```python
def softmax(x):
    total=0
    softmax_arr=[]
    for i in x:
        total=total+np.exp(i)
    for i in x:
        softmax_arr.append(np.exp(i)/total)
    return softmax_arr

x=np.linspace(-10,10,100)
y_softmax=softmax(x)
plt.figure(figsize=(10,6))
plt.plot(x,y_softmax,label="softmax",color="purple")
plt.grid()
plt.title("Softmax activation function")
```
